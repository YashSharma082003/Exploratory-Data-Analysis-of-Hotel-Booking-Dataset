# Hotel Booking Dataset
## Exploratory Data Analysis (EDA)

#### What is EDA? Why do we do EDA?
Exploratory Data Analysis refers to the critical process of performing initial investigations on data to discover patterns, to spot anomalies, test hypotheses and check assumptions with the help of summary statistics and graphical representations.

Exploratory Data Analysis (EDA) is an approach to analysing the data using visual techniques. It is used to discover trends, and patterns, or to check assumptions with the help of statistical summaries and graphical representations.

Exploratory Data Analysis involves two fundamental steps:-
* Data Analysis(Data Preprocessing, Cleaning and Manipulation).
* Data Visualisation(Visualize relationships in data using different types of plots).

#### About Hotel Dataset 
This is an exploratory analysis of [The Hotel Booking Dataset from JESSE MOSTIPAK on Kaggle](https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand?datasetId=511638&language=Python)

Have you ever wondered when the best time of year to book a hotel room is? Or the optimal length of stay in order to get the best daily rate? What if you wanted to predict whether or not a hotel was likely to receive a disproportionately high number of special requests?
This hotel booking dataset can help you explore those questions!

This data set contains booking information for a city hotel and a resort hotel, and includes information such as when the booking was made, length of stay, the number of adults, children, and/or babies, and the number of available parking spaces, among other things.

All personally identifying information has been removed from the data.

The data set that is going to be used is the 'Hotel Booking Dataset'.It contains Hotel Booking data, we are going to clean the data if required and perform Exploratory Data Analysis(EDA). The tools that are going to be used for the EDA would be NumPy, Pandas, Matplotlib and Seaborn.

##### DATA CLEANING AND MODIFICATION 

Fill Null values using `fillna()` function

```python
df[['company','agent','children']] = df[['company','agent','children']].fillna(0)
df['country'].fillna('Not Specified',inplace=True)
```

Add some columns for Data Analysis

```python
df['total_members'] = df['adults'] + df['children'] + df['babies']
df['total_stay'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df['revenue'] = df['total_stay'] * df['adr']
```

Remove duplicate values from our data for Data Analysis

```python
df.drop_duplicates(inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False)[:5]
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

Converting some columns to proper data type and format

```python
df[['children','company','agent']] = df[['children','company','agent']].astype('int64')
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'],format = '%Y-%m-%d')
```

Q1 Which type of Hotel does the guest prefer the most ?

```python
most_prefered_hotel = df['hotel'].value_counts().reset_index().rename(columns={'index':'Hotel','hotel':'Number of Bookings'})
plt.figure(figsize=(10,5))
plt.bar(most_prefered_hotel['Hotel'],most_prefered_hotel['Number of Bookings'],color=['red','blue'])
plt.title("MOST PREFERED HOTEL BY GUEST",fontsize=20)
plt.xlabel("HOTEL",fontsize=10)
plt.ylabel("NUMBERS OF BOOKING",fontsize=10)
plt.show()
```

##### INFERENCES
<pre>
* There are 53428 times City Hotel prefered by guest.
* There are 33968 times Resort Hotel prefered by guest.
<pre>

Q2 What is the percentage of canceled or not canceled of hotel bookings ?

```python
canceled_percentage = df['is_canceled'].value_counts().reset_index().rename(columns={'index':'is_canceled','is_canceled':'Count'})
plt.figure(figsize=(10,5))
plt.pie(x=canceled_percentage['Count'],labels=None,
            data=canceled_percentage,autopct="%.2f%%",explode=(0,0.025))
legend = df['is_canceled'].value_counts().keys()
plt.legend(legend, loc = 'upper right')
plt.title("PERCENTAGE OF CANCELED/NOT CANCELED OF HOTEL BOOKING",fontsize=15)
plt.show()
```

##### INFERENCES
<pre>
* The Hotel Booking were not canceled 63371 times. The Percentage of Not Canceled Hotel booking is 72.51%
* The Hotel Booking were canceled 24025 times. The Percentage of Canceled Hotel booking is 27.49%
<pre>

Q3 Top 10 Countries Earned most Revenue from Hotel Bookings ?

```python
countries_with_most_revenue = df.groupby('country')['revenue'].sum()[:10].reset_index()
plt.figure(figsize=(10,5))
plt.bar(countries_with_most_revenue['country'],countries_with_most_revenue['revenue'],color='lightgreen')
sns.lineplot(x=countries_with_most_revenue['country'],y=countries_with_most_revenue['revenue'],color='black')
plt.title("TOP 10 COUNTRIES EARNED MOST REVENUE FROM HOTEL BOOKINGS",fontsize=20)
plt.xlabel("COUNTRY",fontsize=10)
plt.ylabel("REVENUE",fontsize=10)
plt.show()
```

##### INFERENCES
<pre>
* AGO is the one country who earned most of revenue from hotel bookings.
* The Top 10 Countries who earned most of revenue are AGO, ARG, ARE, AND, ALB, ARM, AIA, ABW, ASM and ATA.
<pre>

Q4 In which year hotels earned more revenue ?

```python
revenue_by_year = df.groupby('arrival_date_year')['revenue'].sum().reset_index()
plt.figure(figsize=(10,5))
plt.pie(x=revenue_by_year['revenue'],labels=revenue_by_year['arrival_date_year'],autopct='%.2f%%')
plt.title("REVENUE EARNED BY HOTELS IN DIFFERENT YEAR",fontsize=15)
plt.show()
```

##### INFERENCES
<pre>
* The hotel earned more revenue in 2016 year. The revenue of year 2016 was 15291176 rupees.
* The hotel earned average revenue in 2017 year. The revenue of year 2016 was 14284246 rupees.
* The hotel earned less revenue in 2015 year. The revenue of year 2016 was 4885517 rupees.
<pre>

Q18 What is the distribution of booking changes made by guests?

## CORRELATION BETWEEN NUMERIC VALUES 

```python
correlation = df.corr()
plt.figure(figsize=(24,12))
sns.heatmap(correlation.corr(),annot=True,fmt = '.2f',cmap='Blues')
plt.title('CORRELATION',fontsize=30)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show();
```

##### INFERENCES
<pre>
* Total_stay and lead time have slight correlation it. This might means that customer plan reservation before their actual arrival.
* Adr(Average Daily Rate) is slightly correlated with total people, which makes sense as more number of people means more revenue.
* Previous booking not cancelled and repeated guest have high correlation as repeated guest are most likely to be who has not cancelled their previous bookings.
* We can also see that some columns have high correlation between them for example total people with adult and children it is because as those column are dervied from them by addition.
